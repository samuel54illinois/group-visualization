```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("building_inventory.csv")

# Drop rows with missing or invalid years
df = df.dropna(subset=["Year Acquired"])
df = df[df["Year Acquired"] > 1800]   # keep only realistic years

# Convert to int
df["Year Acquired"] = df["Year Acquired"].astype(int)

# Create decades
df["Decade"] = (df["Year Acquired"] // 10) * 10

# Count buildings per decade
decade_counts = df["Decade"].value_counts().sort_index()

# Plot
plt.figure(figsize=(12,6))
plt.bar(decade_counts.index, decade_counts.values, width=9, color="skyblue", edgecolor="black")
plt.xlabel("Decade", fontsize=12)
plt.ylabel("Number of Buildings Acquired", fontsize=12)
plt.title("Buildings Acquired by Decade", fontsize=14)

# Rotate and space out ticks
plt.xticks(decade_counts.index[::2], rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
